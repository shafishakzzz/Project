***Lutein and zeaxanthin are carotenoids that are important for eye health. They can help protect the eyes from blue light damage and free radical damage. They may also help delay the onset of age-related macular degeneration (AMD) and cataracts, which are responsible for 56% of blindness globally.***

***Recommended daily allowance: No RDA has been set for either nutrient. Recommended level for eye health: 10 mg/day for lutein and 2 mg/day for zeaxanthin. Safe upper limit: Researchers have not set an upper limit for either. Potential risks: In excess, they may turn your skin slightly yellow***
